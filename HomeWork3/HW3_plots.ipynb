Problem 2:

$$ - \frac{\hbar^2}{2m} \frac{d^2}{dx^2}\psi + V(x)\psi(x) = E\psi(x) $$
$$ \frac{d}{dx}\psi = \phi\frac{d}{dx}\phi = \frac{2m}{\hbar^2}(V(x) - E) $$
$$ Harmonic    Oscillator:       \frac{d}{dx}\psi = \phi\frac{d}{dx}\phi = \frac{2m}{\hbar^2}(V_0 \frac{x^2}{a^2} - E) $$
$$ Anharmonic    Oscillator:       \frac{d}{dx}\psi = \phi\frac{d}{dx}\phi = \frac{2m}{\hbar^2}(V_0 \frac{x^4}{a^4} - E) $$

```python
g_data = pandas.read_csv(str(cwd / "Harmonic_Ground.csv"), header = 0)
f_data = pandas.read_csv(str(cwd / "Harmonic_First.csv"), header = 0)
s_data = pandas.read_csv(str(cwd / "Harmonic_Second.csv"), header = 0)

plt.plot(list(g_data["X"])[:-100], list(g_data["Phi"])[:-100], lw=0.5)
plt.plot(f_data["X"], f_data["Phi"], lw=0.5)
plt.plot(s_data["X"], s_data["Phi"], lw=0.5)

plt.title("V(x) = V0 * x**2 / a**2")
plt.xlabel("x")
plt.ylabel("Phi")
plt.show()
```

```python
ag_data = pandas.read_csv(str(cwd / "Anharmonic_Ground.csv"), header = 0)
af_data = pandas.read_csv(str(cwd / "Anharmonic_First.csv"), header = 0)
as_data = pandas.read_csv(str(cwd / "Anharmonic_Second.csv"), header = 0)


plt.plot(list(ag_data["X"])[300:-300], list(ag_data["Phi"])[300:-300], lw=0.5)
plt.plot(list(af_data["X"])[300:-300], list(af_data["Phi"])[300:-300], lw=0.5)
plt.plot(list(as_data["X"])[300:-300], list(as_data["Phi"])[300:-300], lw=0.5)

plt.title("V(x) = V0 * x**4 / a**4")
plt.xlabel("x")
plt.ylabel("Phi")
plt.show()
```
